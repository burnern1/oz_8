### 리팩토링 과제

Day4에서 만든 "붕어빵 자동 판매 시스템"은 절차적 코드로 구현되어있습니다.
<br>

Day5에서 배운 함수, 클래스 개념을 적용시켜 객체지향적으로 리팩토링을 진행해주세요


```python
class BungeoppangShop:
    def __init__(self):
        # 초기 재고 설정
        self.stock = {
            "팥붕": 10,
            "슈붕": 10,
            "초붕": 10
        }
        # 각 붕어빵 종류별 가격 설정
        self.prices = {
            "팥붕": 1000,
            "슈붕": 1200,
            "초붕": 1500
        }
        # 판매량 초기화
        self.sales = {
            "팥붕": 0,
            "슈붕": 0,
            "초붕": 0
        }

    def check_stock(self):
        # 현재 재고 상태 출력
        for bread_type, bread_count in self.stock.items():
            print(f"종류: {bread_type}, 재고: {bread_count}")
        
    def process_order(self, bread_type, bread_count):
        # 잘못된 붕어빵 종류 체크
        if bread_type not in self.stock:
            print(f"죄송합니다. {bread_type}은(는) 저희 가게에서 판매하지 않습니다.")
            return

        # 재고 확인
        if self.stock[bread_type] < bread_count:
            print(f"죄송합니다. {bread_type}의 재고가 부족합니다.")
            print(f"현재 재고: {self.stock[bread_type]}개")
            return

        # 주문 처리: 재고 감소 및 판매량 증가
        self.stock[bread_type] -= bread_count
        self.sales[bread_type] += bread_count

        # 주문 완료 메시지 출력
        print(f"{bread_type} {bread_count}개 주문이 완료되었습니다.")
        print(f"남은 재고: {self.stock[bread_type]}개")

    def admin_mode(self):
        print("관리자 모드에 진입했습니다.")
        while True:
            # 재고 추가할 붕어빵 종류 입력 받기
            bread_type = input("재고를 추가할 붕어빵 종류를 입력하세요 (종료하려면 'q' 입력): ")
            if bread_type.lower() == 'q':
                print("관리자 모드를 종료합니다.")
                break
            
            # 유효한 붕어빵 종류인지 확인
            if bread_type not in self.stock:
                print(f"오류: {bread_type}은(는) 유효한 붕어빵 종류가 아닙니다.")
                continue
            
            try:
                # 추가할 재고 수량 입력 받기
                additional_stock = int(input(f"{bread_type}의 추가할 수량을 입력하세요: "))
                if additional_stock < 0:
                    print("오류: 음수 값은 입력할 수 없습니다.")
                    continue
                
                # 재고 추가 및 결과 출력
                self.stock[bread_type] += additional_stock
                print(f"{bread_type}의 재고가 {additional_stock}개 추가되었습니다.")
                print(f"현재 {bread_type} 재고: {self.stock[bread_type]}개")
            
            except ValueError:
                print("오류: 올바른 숫자를 입력해주세요.")

    def calculate_total_sales(self):
        # 총 매출 계산
        total_sales = 0
        for item in self.sales:
            total_sales += self.sales[item] * self.prices[item]
        print(f'총 매출: {total_sales:,}원')

def main():
    # BungeoppangShop 객체 생성
    shop = BungeoppangShop()
    
    while True:
        # 메인 메뉴 출력
        print("\n=== 붕어빵 가게 관리 시스템 ===")
        print("1. 주문 모드")
        print("2. 관리자 모드")
        print("3. 재고 확인")
        print("4. 종료")
        
        # 사용자 입력 받기
        choice = input("원하는 모드를 선택하세요 (1-4): ")
        
        if choice == '1':
            # 주문 모드
            bread_type = input("주문할 붕어빵 종류를 입력하세요: ")
            try:
                bread_count = int(input("주문할 수량을 입력하세요: "))
                shop.process_order(bread_type, bread_count)
            except ValueError:
                print("올바른 숫자를 입력해주세요.")
        
        elif choice == '2':
            # 관리자 모드
            shop.admin_mode()
        
        elif choice == '3':
            # 재고 확인
            shop.check_stock()
        
        elif choice == '4':
            # 프로그램 종료
            print("프로그램을 종료합니다.")
            shop.calculate_total_sales()
            break
        
        else:
            print("올바른 선택지를 입력해주세요.")
main()
```

__BungeoppangShop 클래스__

- __init__(): 초기 재고, 가격, 판매량을 설정합니다.
- check_stock(): 현재 재고를 출력합니다.
- process_order(): 주문을 처리하고 재고를 업데이트합니다.
- admin_mode(): 관리자 모드에서 재고를 추가합니다.
- calculate_total_sales(): 총 매출을 계산하고 출력합니다.

목표: main 함수를 작성하여 프로그램의 주요 로직을 실행합니다.

설명:

- "주문" 모드에서는 손님의 주문을 처리하고, "관리자" 모드에서는 재고를 추가할 수 있습니다.
- 프로그램이 종료될 때 총 매출을 계산하고 데이터를 저장합니다.


__[과제 7: 최종 BungeoppangShop 클래스]__

함수로 리팩토링한 "붕어빵 자동 판매 시스템" 작성 후 실행합니다. 

```python
class BungeoppangShop:
    BREAD_TYPES = ["팥붕", "슈붕", "초붕"]

    def __init__(self):
        self.stock = {bread: 10 for bread in self.BREAD_TYPES}
        self.prices = {"팥붕": 1000, "슈붕": 1200, "초붕": 1500}
        self.sales = {bread: 0 for bread in self.BREAD_TYPES}

    def check_stock(self):
        for bread_type, bread_count in self.stock.items():
            print(f"종류: {bread_type}, 재고: {bread_count}")

    def process_order(self, bread_type, bread_count):
        if not self._is_valid_bread_type(bread_type):
            return

        if not self._is_valid_quantity(bread_count):
            return

        if self.stock[bread_type] < bread_count:
            print(f"죄송합니다. {bread_type}의 재고가 부족합니다.")
            print(f"현재 재고: {self.stock[bread_type]}개")
            return

        self.stock[bread_type] -= bread_count
        self.sales[bread_type] += bread_count

        print(f"{bread_type} {bread_count}개 주문이 완료되었습니다.")
        print(f"남은 재고: {self.stock[bread_type]}개")

    def admin_mode(self):
        print("관리자 모드에 진입했습니다.")
        while True:
            bread_type = input("재고를 추가할 붕어빵 종류를 입력하세요 (종료하려면 'q' 입력): ")
            if bread_type.lower() == 'q':
                print("관리자 모드를 종료합니다.")
                break

            if not self._is_valid_bread_type(bread_type):
                continue

            additional_stock = self._get_valid_input(f"{bread_type}의 추가할 수량을 입력하세요: ")
            if additional_stock is None:
                continue

            self.stock[bread_type] += additional_stock
            print(f"{bread_type}의 재고가 {additional_stock}개 추가되었습니다.")
            print(f"현재 {bread_type} 재고: {self.stock[bread_type]}개")

    def calculate_total_sales(self):
        total_sales = sum(self.sales[item] * self.prices[item] for item in self.sales)
        print(f'총 매출: {total_sales}원')

    def _is_valid_bread_type(self, bread_type):
        if bread_type not in self.BREAD_TYPES:
            print(f"오류: {bread_type}은(는) 유효한 붕어빵 종류가 아닙니다.")
            return False
        return True

    def _is_valid_quantity(self, quantity):
        if quantity <= 0:
            print("오류: 올바른 수량을 입력해주세요.")
            return False
        return True

    def _get_valid_input(self, prompt):
        while True:
            try:
                value = int(input(prompt))
                if self._is_valid_quantity(value):
                    return value
            except ValueError:
                print("오류: 올바른 숫자를 입력해주세요.")

def main():
    shop = BungeoppangShop()
    
    while True:
        print("\n=== 붕어빵 가게 관리 시스템 ===")
        print("1. 주문 모드")
        print("2. 관리자 모드")
        print("3. 재고 확인")
        print("4. 종료")
        
        choice = input("원하는 모드를 선택하세요 (1-4): ")
        
        if choice == '1':
            bread_type = input("주문할 붕어빵 종류를 입력하세요: ")
            bread_count = shop._get_valid_input("주문할 수량을 입력하세요: ")
            if bread_count is not None:
                shop.process_order(bread_type, bread_count)
        elif choice == '2':
            shop.admin_mode()
        elif choice == '3':
            shop.check_stock()
        elif choice == '4':
            print("프로그램을 종료합니다.")
            shop.calculate_total_sales()
            break
        else:
            print("올바른 선택지를 입력해주세요.")

if __name__ == "__main__":
    main()
```
