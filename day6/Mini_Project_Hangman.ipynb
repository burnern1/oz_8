```python
import random

# ê²Œì„ì— ì‚¬ìš©ë  ë‹¨ì–´ ëª©ë¡
words = ["apple", "banana", "orange", "grape", "lemon"]

# í–‰ë§¨ ê·¸ë¦¼
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\
     |   / \
     |
    ---"""
]

class HangmanGame:
    def __init__(self):
        # ëœë¤ìœ¼ë¡œ ë‹¨ì–´ ì„ íƒ
        self.word_to_guess = random.choice(words)
        self.guessed_letters = []  # ì‚¬ìš©ìê°€ ì¶”ì¸¡í•œ ê¸€ìë“¤
        self.remaining_attempts = len(hangman_pics) - 1  # ë‚¨ì€ ì‹œë„ íšŸìˆ˜ (í–‰ë§¨ ê·¸ë¦¼ ë‹¨ê³„)
        self.correct_guesses = ["_"] * len(self.word_to_guess)  # ë§ì¶˜ ë‹¨ì–´ ìƒíƒœ

    def display(self):
        # í˜„ì¬ í–‰ë§¨ ìƒíƒœì™€ ë§ì¶˜ ë‹¨ì–´ ìƒíƒœ ì¶œë ¥
        print(hangman_pics[len(hangman_pics) - 1 - self.remaining_attempts])
        print("í˜„ì¬ ë‹¨ì–´: " + " ".join(self.correct_guesses))
        print(f"ì¶”ì¸¡ëœ ê¸€ì: {', '.join(self.guessed_letters)}")
        print(f"ë‚¨ì€ ì‹œë„ íšŸìˆ˜: {self.remaining_attempts}")

    def play(self):
        print("í–‰ë§¨ì— ì˜¤ì‹  ê²ƒì„ í™˜ì˜í•©ë‹ˆë‹¤!!")
        print("ë‹¨ì–´ë¥¼ ì¶”ì¸¡í•´ ë³´ì„¸ìš”!")

        while self.remaining_attempts > 0:
            self.display()
            guess = input("ê¸€ìë¥¼ ë§ì¶°ë³´ì„¸ìš”: ").lower()

            # ì…ë ¥ê°’ ê²€ì¦ (í•œ ê¸€ìë§Œ ì…ë ¥í•´ì•¼ í•¨)
            if len(guess) != 1 or not guess.isalpha():
                print("í•œ ê¸€ìë§Œ ì…ë ¥í•˜ì„¸ìš”")
                continue

            # ì´ë¯¸ ì¶”ì¸¡í•œ ê¸€ìì¸ì§€ í™•ì¸
            if guess in self.guessed_letters:
                print("ì´ë¯¸ ì´ ê¸€ìë¥¼ ë§í˜”ìŠµë‹ˆë‹¤. ë‹¤ì‹œ ì‹œë„í•˜ì„¸ìš”.")
                continue

            # ì¶”ì¸¡í•œ ê¸€ì ì €ì¥
            self.guessed_letters.append(guess)

            # ì •ë‹µ ì—¬ë¶€ í™•ì¸
            if guess in self.word_to_guess:
                print("ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨êµ°ìš”. ë” í˜ë‚´ë´ìš”")
                for idx, letter in enumerate(self.word_to_guess):
                    if letter == guess:
                        self.correct_guesses[idx] = guess  # ë§ì¶˜ ê¸€ì ì—…ë°ì´íŠ¸
            else:
                self.remaining_attempts -= 1
                print(f"í‹€ë ¸ìŠµë‹ˆë‹¤. ë‚¨ì€ ì‹œë„ íšŸìˆ˜: {self.remaining_attempts}")

            # ìŠ¹ë¦¬ ì¡°ê±´ í™•ì¸ (ëª¨ë“  ê¸€ìë¥¼ ë§ì·„ëŠ”ì§€)
            if "_" not in self.correct_guesses:
                print("ì¶•í•˜í•©ë‹ˆë‹¤! ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤")
                print(f"ë‹¨ì–´ëŠ”: {self.word_to_guess}")
                print(f"ë‚¨ì€ ì‹œë„ íšŸìˆ˜ {self.remaining_attempts} ë¡œ ì„±ê³µí–ˆìŠµë‹ˆë‹¤!")
                break

        else:
            # íŒ¨ë°° ì¡°ê±´ (ì‹œë„ íšŸìˆ˜ ì´ˆê³¼)
            print(hangman_pics[-1])
            print(f"ì‹œë„ íšŸìˆ˜ê°€ ë‹¤ ë˜ì—ˆìŠµë‹ˆë‹¤! ë‹¨ì–´ëŠ”: {self.word_to_guess}")

if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
